```python
import json

import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
base_dir = '../output/'
```

```python
df = pd.read_csv(f'{base_dir}df.csv')
df.shape
```

```python
replacements = {}

with open(f'{base_dir}replacements.json', encoding='utf-8') as f:
    replacements = json.load(f)
```

```python
df['key_skill_name'] = df['key_skill_name'].apply(lambda key_skill_name: replacements.get(key_skill_name, key_skill_name))
```

```python
value_counts = df['key_skill_name'].value_counts()
```

```python
valid_values = value_counts[value_counts >= 5].index
```

```python
df = df[df['key_skill_name'].isin(valid_values)].reset_index(drop=True)
df.shape
```

```python
df
```

```python
df.replace({'t': 1, 'f': 0}, inplace=True)
df
```

```python
df = df.dropna().reset_index(drop=True)
df
```
